```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Sample dataset
data = {
    'area': [1000, 1500, 1800, 2400, 3000],
    'bedrooms': [2, 3, 3, 4, 5],
    'bathrooms': [1, 2, 2, 3, 4],
    'price': [100000, 150000, 180000, 240000, 300000]
}

df = pd.DataFrame(data)
df
```

**Step 2: Train the ML Model**

**Step 3: Take User Input & Predict**

```python
# Input from user
area = int(input("Enter area (sqft): "))
bedrooms = int(input("Enter number of bedrooms: "))
bathrooms = int(input("Enter number of bathrooms: "))

# Make prediction
predicted_price = model.predict([[area, bedrooms, bathrooms]])
print(f"\n🏡 Predicted House Price: ₹{int(predicted_price[0])}")
```

```python
import streamlit as st
import pandas as pd
from sklearn.linear_model import LinearRegression

# Sample training data
data = {
    'area': [1000, 1500, 1800, 2400, 3000],
    'bedrooms': [2, 3, 3, 4, 5],
    'bathrooms': [1, 2, 2, 3, 4],
    'price': [100000, 150000, 180000, 240000, 300000]
}
df = pd.DataFrame(data)

# Train the model
X = df[['area', 'bedrooms', 'bathrooms']]
y = df['price']
model = LinearRegression()
model.fit(X, y)

# Streamlit UI
st.set_page_config(page_title="🏠 House Price Predictor", layout="centered")

st.title("🏡 House Price Prediction App")
st.markdown("Enter house details to predict the price:")

# User inputs
area = st.number_input("Area (in sqft)", min_value=500, max_value=10000, step=100)
bedrooms = st.slider("Number of Bedrooms", 1, 10, 3)
bathrooms = st.slider("Number of Bathrooms", 1, 5, 2)

if st.button("Predict Price"):
    input_data = pd.DataFrame([[area, bedrooms, bathrooms]], columns=['area', 'bedrooms', 'bathrooms'])
    prediction = model.predict(input_data)
    st.success(f"🏷 Estimated House Price: ₹{int(prediction[0]):,}")
```
